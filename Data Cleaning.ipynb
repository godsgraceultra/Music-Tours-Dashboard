```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\archive (40)\my_file (1).csv")
df.head()
```

```python
df.info()
```

```python
df= df.drop(columns = ["Rank","Peak","All Time Peak","Ref."])
df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df.rename(columns = {"Actual\xa0gross":"Actual Gross","Adjusted\xa0gross (in 2022 dollars)":"Adjusted gross (in 2022 dollars)"})
df.head(1)
```

```python
df[["Actual Gross","Adjusted gross (in 2022 dollars)","Average gross", "Tour title"]]
```

```python
df["Actual Gross"] = df["Actual Gross"].str.replace(r"[\$,]", "", regex=True)  
df["Actual Gross"] = df["Actual Gross"].str.replace(r"\[.*\]", "", regex=True)  
df["Adjusted gross (in 2022 dollars)"] = df["Adjusted gross (in 2022 dollars)"].str.replace(r"[\$,]", "", regex=True)
df["Average gross"] = df["Average gross"].str.replace(r"[/$,]", "", regex=True)
df["StartYear"] = df["Year(s)"].str.extract(r"(\d{4})")
df["Tour title"] = df["Tour title"].str.replace(r"[†‡\*]", "", regex=True)
df["Tour title"] = df["Tour title"].str.replace(r"\[.*?\]", "", regex=True)
df = df.drop(columns=["Year(s)"])
df.head(2)
```

```python
df["Actual Gross"] = pd.to_numeric(df["Actual Gross"])
df["Adjusted gross (in 2022 dollars)"] = pd.to_numeric(df["Adjusted gross (in 2022 dollars)"])
df["StartYear"] = pd.to_numeric(df["StartYear"])
df["Average gross"] = pd.to_numeric(df["Average gross"])
df["Tour title"] = df["Tour title"].str.strip()
df.head()
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
print("File Saved Susscessfully")
```
